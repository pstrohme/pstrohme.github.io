```python
import matplotlib.pyplot as plt
import numpy as np

# Beispiel-Daten
np.random.seed(0)
data = np.concatenate([np.random.normal(loc=-2, scale=1, size=50),
                       np.random.normal(loc=2, scale=1, size=50),
                       np.random.normal(loc=5, scale=2, size=5)])

plt.figure(figsize=(8,3))

# Boxplot erstellen
bp = plt.boxplot(data, sym='o', vert=False, patch_artist=True, widths=0.09)

# Farben für die Bereiche festlegen
box_color = 'lightgray'
whisker_color = 'purple'
flier_color = 'darkred'
median_color = 'green'
edge_color = 'darkblue'

# Bereiche einfärben
for patch in bp['boxes']:
    patch.set_facecolor(box_color)
    patch.set_edgecolor(edge_color)
for patch in bp['whiskers']:
    patch.set(color=whisker_color, linewidth=1.5)
for patch in bp['fliers']:
    patch.set(marker='o', color=flier_color)
for patch in bp['medians']:
    patch.set(color=median_color, linewidth=2)

# Beschriftungen für Median, Quartile und Whisker hinzufügen
plt.text(np.median(data), 1.075, 'Median', ha='center', color=median_color)
plt.text(np.percentile(data, 25), 1.075, 'Q1', ha='center', color=edge_color)
plt.text(np.percentile(data, 75), 1.075, 'Q3', ha='center', color=edge_color)
plt.text(np.min(data), 1.05, 'unterer \n "Whisker"', ha='center', color=whisker_color)
plt.text(np.max(data[data <= np.percentile(data, 75) + 1.5 * (np.percentile(data, 75) - np.percentile(data, 25))]), 1.05, 'oberer \n "Whisker"', ha='center', color=whisker_color)
plt.text(np.max(data), 0.95, 'Ausreißer', ha='center',color=flier_color)

# Entferne die obere und rechte Achsenlinie
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# Setze die Achsentick-Positionierung
plt.gca().tick_params(axis='both', which='both', direction='out')
plt.gca().xaxis.set_tick_params(width=0.5)
plt.gca().yaxis.set_tick_params(width=0.5)

plt.xlim(-6, 10)
# Y-Achse verschieben
plt.ylim(0.8, 1.2)

# Anzeige des Boxplots
plt.show()
```

```python
import numpy as np

val_accuracy = [0.85, 0.83, 0.81, 0.8, 0.78, 0.75, 0.72, 0.7, 0.65]
threshold = [0.6] * len(val_accuracy)

# Zeitpunkte
start_date = '2023-05-10'
end_date = '2023-06-30'
dates = pd.date_range(start=start_date, end=end_date, periods=len(val_accuracy))

# Plot für den Verlauf der Genauigkeit
plt.plot(dates, val_accuracy, 'k-', label='Test-Genauigkeit')
plt.plot(dates, threshold, 'r--', label="Threshold für Re-Training")
plt.title('Modellverfall - Genauigkeit')
plt.xlabel('Datum')
plt.ylabel('Genauigkeit')
plt.ylim(0,1)
plt.legend()
plt.xticks(rotation=45)
plt.savefig("../Grafiken/Modelverfall.png")
```
